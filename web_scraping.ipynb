```python
price = soup.find("h4",  {"class":"float-end price card-title pull-right"})
```

```python
print(price.get_text())
```

```python
allPrices = soup.findAll("h4", {"class": "float-end price card-title pull-right"})
```

```python
print(allPrices)
```

```python
for price in allPrices:
    print(price.get_text())
```

```python
items = soup.findAll("a", {"class": "title"})
```

```python
print(items)
```

```python
count = 0
for item in items:
    print(item.get_text())
    count = count + 1
```
